```python
from mpl_toolkits.basemap import Basemap
import pandas as pd

m = Basemap(llcrnrlon = 30, llcrnrlat = -35, urcrnrlon = 65, urcrnrlat = -10, projection = 'merc',resolution ='l',area_thresh=1000.)
m.drawcoastlines(color = 'black', linewidth=1.5)

df = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/track.csv')
#dt['x'], dt['a'] = m(list(dt['Longitude']), list(dt['Latitude']))
m(list(df['Longitude1']), list(df['Latitude1']))
#m(list(df['Longitude2']), list(df['Latitude2']))
m(list(df['Longitude3']), list(df['Latitude3']))
m(list(df['Longitude4']), list(df['Latitude4']))
m(list(df['Longitude5']), list(df['Latitude5']))

df['x1'], df['y1']=m(list(df['Longitude1']), list(df['Latitude1']))
#df['x2'], df['y2']=m(list(df['Longitude2']), list(df['Latitude2']))
df['x3'], df['y3']=m(list(df['Longitude3']), list(df['Latitude3']))
df['x4'], df['y4']=m(list(df['Longitude4']), list(df['Latitude4']))
df['x5'], df['y5']=m(list(df['Longitude5']), list(df['Latitude5']))


#plt.plot(dt['x'], dt['a'], marker='v',color='m', lw = 3)

#m.drawmapboundary(fill_color='white')
#m.fillcontinents(color='white')

#m.drawcoastlines()
m.plot(df['x1'], df['y1'],'o-',color='blue',label = 'Sim1')
#m.plot(df['x2'], df['y2'],'o-',color='red',label = 'Sin2')
m.plot(df['x3'], df['y3'],'o-',color='green',label = 'Sim3')
m.plot(df['x4'], df['y4'],'o-',color='black',label = 'Sim4')
m.plot(df['x5'], df['y5'],'o-',color='red',label = 'Obs')
m.drawcountries()
m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966',lake_color='#99ffff')
m.drawparallels(np.arange(10,70,20),labels=[1,1,0,0])
m.drawmeridians(np.arange(-100,0,20),labels=[0,0,0,1])
plt.title('Tropical cyclone track')
m.drawparallels(np.arange(30,65,5),labels=[1,1,0,1], color='black', linewidth=2.0)
m.drawmeridians(np.arange(-35,-10,5),labels=[1,1,0,1], color='black', linewidth=2.0)
plt.legend()
#plt.title('Track of tropical cyclone')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#plt.box()
#plt.grid()
plt.show()
#plt.grid()
#filout= "track.png"
#fig.savefig('tr',dpi=300)
#plt.savefig('track.png')
```
