```python
latlon = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/latlon.csv')
observation = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/observation.csv')
sim3 = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/sim3.csv')
sim4 = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/sim4.csv')
sim2 = pd.read_csv('/home/aimsstudent45/Downloads/Essay/Paper/sim2.csv')
```

```python
x=latlon['Longitude']
y=latlon['Latitude']
x1=observation['Longitude']
y1=observation['Latitude']
x2=sim3['Longitude']
y2=sim3['Latitude']
x3=sim4['Longitude']
y3=sim4['Latitude']
x4=sim2['Longitude']
y4=sim2['Latitude']
plt.figure(figsize=(10,10))
plt.plot(x, y,'o-',color='blue',label = 'Sim1')
plt.plot(x1, y1,'o-',color='red',label = 'Observed')
plt.plot(x2, y2,'o-',color='black',label = 'sim3')
#plt.plot(x3, y3,'o-',color='',label = 'sim4')
plt.plot(x4, y4,'o-',color='green',label = 'sim2')
plt.xlabel('Longitude',fontsize=15)
plt.ylabel('Latitude',fontsize=15)
plt.title('Track of Tropical cyclone Enawo')
plt.grid()
plt.legend()
plt.savefig('track.png') 
plt.show()
plt.close()
```
